```python
import pandas as pd

# Load your fixed dataset
df = pd.read_excel("career_dataset_fixed.xlsx")

def get_main_category(career):
    # Ensure it's a string and lowercased for accurate matching
    c = str(career).lower().strip()
    
    # Priority-based clustering
    if "ai " in c or "artificial intelligence" in c: 
        return "Artificial Intelligence"
    elif "ios" in c or "android" in c or "mobile" in c: 
        return "Mobile App Development"
    elif "cloud" in c or "aws" in c or "azure" in c: 
        return "Cloud Computing"
    elif "cyber" in c or "security" in c: 
        return "Cybersecurity"
    elif "data" in c or "analytics" in c: 
        return "Data Science & Analytics"
    elif "devops" in c: 
        return "DevOps"
    elif "ui" in c or "ux" in c or "design" in c: 
        return "UI/UX Design"
    elif "backend" in c or "java" in c: 
        return "Backend Development"
    elif "frontend" in c or "web" in c: 
        return "Web Development"
    elif "game" in c:
        return "Game Development"
    else:
        return "General IT Engineering"

# Apply mapping
df['Main_Category'] = df['Career'].apply(get_main_category)

# Save to a new clustered file
df.to_excel("career_dataset_clustered.xlsx", index=False)

# --- VERIFICATION STEP ---
# Print the count to see how many items fell into 'General IT Engineering'
# This helps you decide if you need to add another category!
print("Clustering Complete.")
print(df['Main_Category'].value_counts())

# If you want to see which specific titles went to 'General IT Engineering':
general_list = df[df['Main_Category'] == "General IT Engineering"]['Career'].unique()
print("\nTitles currently in 'General IT Engineering':")
print(general_list[:10]) # Printing first 10 to check
```
